### Buat DataFrame yang mendeskripsikan Tiap Tiap kolom dari data Startup Funding

```python
datafeat,datatype,null,nullpct,unique,uniquesamp = [],[],[],[],[],[]
for item in df.columns:
    datafeat.append(item)
    datatype.append(df[item].dtype)
    null.append(df[item].isnull().sum())
    nullpct.append(round(((df[item].isnull().sum() / len(df))*100),2))
    unique.append(len(df[item].unique()))
    uniquesamp.append(list(df[item].value_counts().head(2).index))

dictx = {
    'dataFeatures' : datafeat,
    'dataType' : datatype,
    'null' : null,
    'nullPct' : nullpct,
    'unique' : unique,
    'uniqueSample' : uniquesamp
}

pd.DataFrame(dictx)
```
